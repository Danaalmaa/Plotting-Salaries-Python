```python
import matplotlib.pyplot as plt
import pandas as pd
ndf = pd.read_csv('salaries.csv', delimiter = ',')
sColumn = pd.DataFrame(ndf, columns=["salary"])
hh= sColumn.sort_values(by=['salary'], ascending=True)
display(hh)
hh.plot.line()
```

```python
#using numpy
import numpy as np
remote_arr = np.array(ndf['remote_ratio'])
remote_labels, remote_counts = np.unique(ndf['remote_ratio'],return_counts=True)
plt.figure()
lables_remote = ['NO REMOTE WORK','PARTIALLY REMOTE','FULLY REMOTE']
plt.pie(remote_counts, labels = lables_remote,autopct='%.2f%%')
plt.show()
```
